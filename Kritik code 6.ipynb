```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

def normal_density(mean, variance, x):
    """Returns the value of the normal density at x."""
    return 1/np.sqrt(2*np.pi*variance) * np.exp(-((x-mean)**2)/(2*variance))

def integration(mean, variance, a, b):
    """Numerically integrates the normal density between a and b."""
    result, error = quad(lambda x: normal_density(mean, variance, x), a, b)
    return result

# Example: Plot the density for two different sets of parameters.
x_vals = np.linspace(150, 200, 500)
plt.figure()
plt.plot(x_vals, normal_density(171, 7.12, x_vals), label='μ = 171, σ² = 7.12')
plt.plot(x_vals, normal_density(160, 10, x_vals), label='μ = 160, σ² = 10')
plt.xlabel("x")
plt.ylabel("f_X(x)")
plt.title("Normal Density Functions")
plt.legend()
plt.show()

# Global male heights: μ = 171 cm and variance σ² = 7.12.
# Determine the probability that a randomly chosen man has height between 162 and 190 cm.
probability = integration(171, 7.12, 162, 190)
print("Probability (height between 162 cm and 190 cm):", probability)
```

```python
#Alternatively, we can compute the expected dosage numerically by “recycling” the code from Problem 3. For example:
def expected_dosage(mean, variance):
    """Numerically compute the expected dosage given D(x)=2.38*x^2."""
    std = np.sqrt(variance)
    # Integrate over a range covering most of the probability mass
    a = mean - 5*std
    b = mean + 5*std
    result, error = quad(lambda x: 2.38*x**2 * normal_density(mean, variance, x), a, b)
    return result

print("Expected dosage:", expected_dosage(171, 7.1))
```
